```python
from LLmsfJiT import read_trec_qrels, get_tables_from_qrels, connect, config
import pandas as pd
```

```python
params = config("../database.ini")
conn, cur = connect(params["postgres"])
```

```python
qrels = read_trec_qrels("../rel_files/rel_table_qrels.txt")
tables = [t.table for t in get_tables_from_qrels(conn, cur, "../rel_files/rel_table_qrels.txt")]
```

```python
table_df = pd.DataFrame(tables)
```

```python
exit_state = 0
counter = 0
while exit_state == 0:
    t = table_df.loc[(table_df["has_header"] == True) & (table_df.relation.map(len)==5)].iloc[counter]
    print("\033[F")
    print(counter)
    display(pd.DataFrame(t.relation))
    if input() == "exit":
        exit_state = 1
    else:
        counter += 1
```
